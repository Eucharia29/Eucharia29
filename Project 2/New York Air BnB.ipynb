# Data Cleaning

```python
# converting data type and removing NaN from latitude 

df['latitude'] = pd.to_numeric(df['latitude'],errors='coerce')
df['latitude'] = df['latitude'].fillna(df['latitude'].mean())

print(df[['latitude','longitude']].isnull().sum())
print(df['latitude'].tail())
print(df['longitude'].tail())
```

```python
# converting data type and removing NaN in Last_Review 
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

df['last_review'] = pd.to_datetime(df['last_review'],dayfirst=True,errors='coerce')
#df['last_review'] = df['last_review'].where(df['last_review'].notna(),"No Reviews")

print(df[['last_review', 'reviews_per_month']].isnull().sum())
print(df['last_review'].head())
```

```python
object_columns = df.select_dtypes(include=['object']).columns
print(f"Object Colunms: {object_columns}")

```

```python
#object column cleaning
#import re
for col in object_columns:
    df[col] = df[col].astype(str)

def clean_text(text):
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
        text = text.strip().title()
        return text


for col in object_columns:
    df[col] = df[col].apply(clean_text)

df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
```
